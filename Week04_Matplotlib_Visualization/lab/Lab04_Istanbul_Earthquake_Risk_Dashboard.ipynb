## Hints

### Gutenberg-Richter Fit

To compute the Gutenberg-Richter relationship:

```python
# Sort unique magnitudes
mags = np.sort(df['magnitude'].values)
# For each magnitude M, count earthquakes >= M
unique_mags = np.arange(mags.min(), mags.max(), 0.1)
cumulative_counts = np.array([np.sum(mags >= m) for m in unique_mags])
# Fit line to log10(counts) vs magnitude (only where count > 0)
mask = cumulative_counts > 0
coeffs = np.polyfit(unique_mags[mask], np.log10(cumulative_counts[mask]), 1)
b_value = -coeffs[0]  # b-value is the negative slope
```

### Colormaps

**Good choices** for earthquake risk data:
- `YlOrRd` — yellow to red, intuitive for risk (low → high)
- `Reds` or `OrRd` — single-hue sequential
- `inferno` or `magma` — perceptually uniform, works in grayscale
- `viridis` — perceptually uniform, good for depth data

**Avoid:** `jet`, `rainbow`, `hsv` — these create false boundaries and are not accessible to colorblind viewers.

### GridSpec Layout

```python
fig = plt.figure(figsize=(16, 10))
gs = fig.add_gridspec(2, 3, hspace=0.3, wspace=0.3)

ax1 = fig.add_subplot(gs[0, :2])   # top-left, spans 2 columns

```

> *"Above all else, show the data."* — Edward Tufte

Remove unnecessary gridlines, borders, and decorations. Every drop of ink on your chart should represent data or help the reader interpret data. If an element doesn't serve either purpose, remove it.

### Deliverable 3: Magnitude-Frequency Analysis

````python
# --- Deliverable 3: Magnitude–frequency ---

mag = df_quakes["magnitude"].dropna().values
fig, axes = plt.subplots(1, 2, figsize=(12, 4.5))

# Subplot 1: histogram + KDE (bin width 0.2 — catalog often reported to 0.1 M; smooths noise)
bins = np.arange(mag.min(), mag.max() + 0.21, 0.2)
sns.histplot(mag, bins=bins, kde=True, stat="density", color="steelblue", ax=axes[0], edgecolor="white")
mean_m, med_m = np.mean(mag), np.median(mag)
axes[0].axvline(mean_m, color="darkred", ls="--", lw=1.5, label=f"Mean = {mean_m:.2f}")
axes[0].axvline(med_m, color="darkgreen", ls=":", lw=1.5, label=f"Median = {med_m:.2f}")
axes[0].set_xlabel("Magnitude")
axes[0].set_ylabel("Density")
axes[0].set_title("Magnitude distribution (histogram + KDE)")
axes[0].legend()

# Subplot 2: Gutenberg–Richter cumulative (log10 N vs M)
m_sorted = np.sort(mag)
step = 0.1
mag_edges = np.arange(np.floor(m_sorted.min() * 10) / 10, np.ceil(m_sorted.max() * 10) / 10 + step, step)
cum = np.array([np.sum(m_sorted >= m) for m in mag_edges])
mask = cum > 5  # avoid log of tiny tail noise
x_fit, y_fit = mag_edges[mask], np.log10(cum[mask])
coeffs = np.polyfit(x_fit, y_fit, 1)
b_value = -coeffs[0]

axes[1].plot(mag_edges, np.log10(np.maximum(cum, 1)), color="black", lw=1.2, label="log₁₀ N(≥M)")
axes[1].plot(
    x_fit,
    np.polyval(coeffs, x_fit),
    color="crimson",
    ls="--",
    lw=1.5,
    label=f"Linear fit (b ≈ {b_value:.2f})",
)
axes[1].set_xlabel("Magnitude M")
axes[1].set_ylabel("log₁₀(cumulative count), N(≥M)")
axes[1].set_title("Gutenberg–Richter (cumulative)")
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
````

````python
# --- Deliverable 6A: Multi-panel executive dashboard ---

merged = df_districts.merge(df_hazard[["district", "pga_g"]], on="district", how="left")
merged["exposure_index"] = (
    (merged["population"] / merged["population"].max())
    * merged["pga_g"]
    * (merged["pre1999_building_pct"] / 100.0)
)
merged["exposure_index"] /= merged["exposure_index"].max()

fig = plt.figure(figsize=(16, 10))
gs = fig.add_gridspec(2, 3, hspace=0.36, wspace=0.32)

# Top row: timeline (2 cols) + derived metric bars
ax_timeline = fig.add_subplot(gs[0, :2])
sc_t = ax_timeline.scatter(
    df_quakes["year"],
    df_quakes["magnitude"],
    c=df_quakes["depth_km"],
    cmap="viridis",
    s=np.clip(df_quakes["magnitude"] ** 2.5 * 3, 6, 200),
    alpha=0.55,
    edgecolors="white",
    linewidths=0.2,
)
fig.colorbar(sc_t, ax=ax_timeline, label="Depth (km)")
ax_timeline.set_xlabel("Year")
ax_timeline.set_ylabel("Magnitude")
ax_timeline.set_title("Historical Marmara seismicity (size ∝ magnitude)")
ax_timeline.grid(True, alpha=0.25)

ax_exp = fig.add_subplot(gs[0, 2])
top_ex = merged.nlargest(10, "exposure_index").sort_values("exposure_index")
ax_exp.barh(top_ex["district"], top_ex["exposure_index"], color="darkred")
ax_exp.set_xlabel("Normalized exposure index")
ax_exp.set_title("Top districts: pop × hazard × older stock")
for l in ax_exp.get_yticklabels():
    l.set_fontsize(8)

# Bottom: histogram + spatial hazard (spans 2 cols)
ax_hist = fig.add_subplot(gs[1, 0])
ax_hist.hist(
    df_quakes["magnitude"].dropna(),
    bins=np.arange(2, np.ceil(df_quakes["magnitude"].max()) + 0.5, 0.25),
    color="steelblue",
    edgecolor="white",
)
ax_hist.set_xlabel("Magnitude")
ax_hist.set_ylabel("Count")
ax_hist.set_title("Catalog magnitude frequency")

ax_map = fig.add_subplot(gs[1, 1:])
h_map = df_hazard.merge(merged[["district", "population"]], on="district")
sc_m = ax_map.scatter(
    h_map["longitude"],
    h_map["latitude"],
    c=h_map["pga_g"],
    s=h_map["population"] / 3500,
    cmap="YlOrRd",
    alpha=0.9,
    edgecolors="0.25",
    linewidths=0.4,
)
fig.colorbar(sc_m, ax=ax_map, label="Relative hazard index")
ax_map.set_xlabel("Longitude (°E)")
ax_map.set_ylabel("Latitude (°N)")
ax_map.set_title("District hazard (color) and population (marker size)")
ax_map.set_aspect("equal", adjustable="box")
ax_map.grid(True, alpha=0.2)

fig.suptitle(
    "Istanbul earthquake risk — council briefing: hazard, exposure, and catalog context",
    fontsize=15,
    fontweight="bold",
)

out_path = LAB_DIR / "dashboard.png"
fig.savefig(out_path, dpi=300, bbox_inches="tight")
plt.show()
print("Saved:", out_path.resolve())
````

#### Reflection

**1. Chart type justification (D2–D5):**

- **D2 (scatter timeline):** I used **position** (year vs. magnitude) as the highest-accuracy encoding channel from the theory outline, added **size** for magnitude emphasis and **color** for depth, plus a **sequential** colormap (`viridis`) so depth reads as ordered magnitude, not categorical noise.
- **D3 (histogram + G–R):** The histogram respects the **bin-width** discussion in the notes; the cumulative **`log₁₀` transform** makes the power-law (Gutenberg–Richter) segment visible—the same idea as “right transformation reveals structure.”
- **D4 (sorted bars):** For **categorical comparisons** with a clear ordering (population, vulnerability proxy), **bar charts** beat pie charts and keep **data–ink** high; sorting by substance rather than alphabet follows the chart-selection guidance.
- **D5 (map-style scatter):** **Spatial data → spatial chart**; `YlOrRd` is **sequential** and avoids **jet/rainbow** pitfalls; annotations anchor **Fatih / Esenler–class** highs and geographic labels.

**2. Audience adaptation (seismologists vs. municipal council):**

- **Example 1:** For seismologists I would show **depth histograms by source**, **magnitude type** (Mw vs. local), **completeness** notes, and **G–R residuals**; the council version hides that in favor of **event counts** and **plain-language** titles.
- **Example 2:** I would replace illustrative **`pga_g` indices** with **explicit g-levels** or **code-based spectrum** parameters and show **uncertainty**, whereas elected officials need **ranked districts** and **budget-friendly** summaries.

**3. Self-critique:**

The **weakest** panel is the **district hazard scatter (D5)** because the **`pga_g` column is an ordered index, not a verified AFAD grid**. I would improve it by **importing official PGA / soil maps**, using a **contourf** or **choropleth on district boundaries**, and documenting **datum and CRS**—avoiding **false precision** is called out in the theory notes as a common integrity issue.
